```python
import pandas as pd
import numpy as np
```

```python
DATASET_FILE = "./dataset_cleaned.csv"
```

```python
dataset_df = pd.read_csv(DATASET_FILE)
dataset_df
```

```python
# having a datafarme containing negative data

Negative_data = dataset_df[(dataset_df['stars'] == 1) | (dataset_df['stars'] == 2)]

Neg_DF = pd.DataFrame(columns=['text_cleaned','stars','length'])
Neg_DF.text_cleaned= Negative_data.text_cleaned
Neg_DF.stars= Negative_data.stars
Neg_DF.length= Negative_data.length
Neg_DF.index=range(len(Negative_data))
print(Neg_DF)
```

**vectorize the preprocessed dataset.**

```python
from sklearn.feature_extraction.text import TfidfVectorizer

corpus = Neg_DF.text_cleaned
vectorizer = TfidfVectorizer(max_df=0.8, min_df=0.01)
X = vectorizer.fit_transform(corpus)
vectorizer.get_feature_names_out()
#10 000
print(X.shape)
print(X)
```

```python
# Create data-frame of Doc-Term Matrix with review text as column names
data_dtm_text = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names_out())
# Set President's Names as Index
data_dtm_text.index = Neg_DF.index
# Visually inspect Document Term Matrix
data_dtm_text.head()
```

**train the topic modeling model**

```python
#Display Topics 
def display_topics(model, feature_names, num_top_words,topic_names=None):
    # iterate through topics in topic-term matrix, 'H' aka
    # model.components_
  for ix, topic in enumerate(model.components_):

  #print topic, topic number, and top words
    if not topic_names or not topic_names[ix]:
      print("\nTopic ", ix)
    else:
      print("\nTopic: '",topic_names[ix],"'")
    print(", ".join([feature_names[i] \
    for i in topic.argsort()[:-num_top_words - 1:-1]]))
  
  
  
  

            
        
            
        
        


```

```python
from sklearn.decomposition import NMF
```

```python
nmf_model = NMF(15)
doc_topic = nmf_model.fit_transform(data_dtm_text)
display_topics(nmf_model, vectorizer.get_feature_names(), 5)
```

```python
#Dictionnaire de topic en fonction de son theme
mon_dictionnaire={0:"Unadmirable place",1:"Insufficient quantity of meal",2:"Pizza",3:"Wrong delivery",4:"Slow service",5:"Service staff",6:"burger dish",7:"waiting time",8:"chicken dish",9:"evening",10:"location",11:"Unsatisfied customer",12:"Full menu request",13:"seafood menu request",14:"wonderful meal"}
print(mon_dictionnaire)
```

```python
#Conversion du matrice de correlation des reviews avec les 15 topics en dataframe
df_doc_topic = pd.DataFrame(doc_topic)
print(df_doc_topic)
```
